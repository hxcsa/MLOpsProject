```python
df = pd.read_csv("../artifacts/raw/data.csv")
y = df["Survival_Prediction"]
x = df.drop(columns=["Survival_Prediction"])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
caregorical_cols = x.select_dtypes(include = ['object']).columns
label_encoders = {}

for col in caregorical_cols:
    le = LabelEncoder()
    x[col] = le.fit_transform(x[col])
    label_encoders[col] = le
```

```python
x.head
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x ,y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_selection import chi2, SelectKBest

x_cat = x_train.select_dtypes(include=['int64', 'float64'])
chi2selector = SelectKBest(score_func=chi2, k = 'all')
chi2selector.fit(x_cat, y_train)
```

```python
chi2_score = pd.DataFrame({
    'Feature' : x_cat.columns,
    "chi2_score" : chi2selector.scores_
}).sort_values(by = 'chi2_score', ascending=False)
```

```python
top_features = chi2_score.head(5)["Feature"].tolist()
```

```python
x =x[top_features]
```
